```python
#CREA ESQUEMAS DE DATOS
import pyspark.sql.functions as F
from pyspark.sql.types import *
#ESQUEMA PARA BROKER
schema_broker = StructType([
    StructField("id_broker", IntegerType(), False),
    StructField("nombre", StringType(), True),
    StructField("region", StringType(), True),
    StructField("email", StringType(), True)
])

#ESQUEMA PARA CAMPAINGS
schema_campania = StructType([
    StructField("id_campania", IntegerType(), False),
    StructField("canal", StringType(), True),
    StructField("nombre", StringType(), True),
    StructField("fecha_emision", DateType(), True),
    StructField("fecha_fin", DateType(), True),
    StructField("presupuesto", DoubleType(), True)
])

#ESQUEMA PARA CLIENTES
schema_cliente = StructType([
    StructField("id_cliente", IntegerType(), False),
    StructField("nombre", StringType(), True),
    StructField("apellido", StringType(), True),
    StructField("email", StringType(), True),
    StructField("region", StringType(), True)
])

#ESQUEMA PARA INTERESADOS
schema_interesado = StructType([
    StructField("id_interesado", IntegerType(), False),
    StructField("id_cliente", IntegerType(), False),
    StructField("id_propiedad", IntegerType(), True),
    StructField("id_campania", IntegerType(), True),
    StructField("fecha_emision", DateType(), True),
    StructField("medio", StringType(), True)
])

#ESQUEMA PARA PROYECTOS
schema_proyecto = StructType([
    StructField("id_proyecto", IntegerType(), False),
    StructField("nombre", StringType(), True),
    StructField("ciudad", StringType(), True),
    StructField("region", StringType(), True),
    StructField("fecha_lanzamiento", DateType(), True),
    StructField("estado", StringType(), True)
])

#ESQUEMA PARA PROPIEDADES
schema_propiedad = StructType([
    StructField("id_propiedad", IntegerType(), False),
    StructField("id_proyecto", IntegerType(), False),
    StructField("tipo", StringType(), True),
    StructField("tamanio_m2", DoubleType(), True),
    StructField("habitaciones", IntegerType(), True),
    StructField("banios", IntegerType(), True),
    StructField("precio", DoubleType(), True),
    StructField("estado", StringType(), True)
])

#ESQUEMA PARA VENTAS
schema_ventas = StructType([
    StructField("id_venta", IntegerType(), False),
    StructField("id_propiedad", IntegerType(), False),
    StructField("id_cliente", IntegerType(), True),
    StructField("id_broker", IntegerType(), True),
    StructField("fecha_venta", DateType(), True),
    StructField("precio_venta", DoubleType(), True)
])

print("✅ Esquemas definidos correctamente")
```

```python
# Leer los archivos CSV con los esquemas definidos anteriormente
try:
    # Leer archivo de brokers
    df_brokers = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_broker) \
        .load("Files/raw/brokers.csv")
    
    # Leer archivo de campanias
    df_campanias = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_campania) \
        .load("Files/raw/campaigns.csv")
    
    # Leer archivo de clientes
    df_clientes = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_cliente) \
        .load("Files/raw/clients.csv")
    
    # Leer archivo de interesados
    df_interesados = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_interesado) \
        .load("Files/raw/leads.csv")
    
    # Leer archivo de proyectos
    df_proyectos = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_proyecto) \
        .load("Files/raw/projects.csv")
    
    # Leer archivo de propiedades
    df_propiedades = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_propiedad) \
        .load("Files/raw/properties.csv")

    # Leer archivo de ventas
    df_ventas = spark.read.format("csv") \
        .option("header", "true") \
        .schema(schema_ventas) \
        .load("Files/raw/sales.csv")
    print("✅ Archivos CSV leídos correctamente")
except Exception as e:
    print(f"Error al leer archivos: {str(e)}")
```

```python
# Aplicamos transformaciones para mejorar la calidad de los datos
# Transformación de broker: 
#quita valores null
#formatea campos de texto
df_broker_procesado = df_brokers.dropna(subset=["id_broker", "nombre"]) \
.withColumn("region", F.upper("region")) \


# Transformaciones de campanias:
# formatea campos texto
df_campanias_procesado = df_campanias.dropna(subset=["id_campania", "nombre"]) \
    .withColumn("canal", F.lower(F.col("canal"))) 

# Transformación de venta:
# Quita valores null
# Extraer el año de venta
# Extraer el mes de venta
df_ventas_procesado = df_ventas.dropna(subset=["id_propiedad", "id_cliente", "fecha_venta"]) \
.withColumn(
    "anio_venta", 
    F.year(F.col("fecha_venta"))) \
.withColumn(
    "mes_venta", 
    F.month(F.col("fecha_venta")))

# Transformación de clientes:
# Añadir columna nombre_completo
# eliminar duplicados
df_clientes_clean= df_clientes.withColumn(
    "nombre_completo", 
    F.concat(F.col("nombre"), F.lit(" "), F.col("apellido"))) \
    .dropDuplicates(["id_cliente"]) \
    .withColumn("region", F.upper("region")) \

#Compras que ha tenido el cliente
# Paso 1: Contar las compras por cliente
compras_por_cliente = df_ventas_procesado.groupBy("id_cliente") \
    .agg(F.count("id_venta").alias("numero_compras"))
#  Unir esa información al DataFrame de clientes
df_cliente_con_compras = df_clientes_clean.join(compras_por_cliente, on="id_cliente", how="left")
# los clientes sin compras tengan 0 en lugar de null
df_clientes_procesado = df_cliente_con_compras.fillna({"numero_compras": 0})


# Transformaciones de interesados:
# formatea campos texto
df_interesados_procesado = df_interesados.dropna(subset=["id_cliente", "id_propiedad"]) \
    .withColumn("medio", F.lower(F.col("medio"))) 


# Transformación de proyecto: 
#quita valores null
#Extraer el año de lanzamiento
#formatea campos de texto
df_proyectos_procesado = df_proyectos.dropna(subset=["id_proyecto", "nombre"]) \
.withColumn(
    "anio_lanzamiento", 
    F.year(F.col("fecha_lanzamiento")))  \
.withColumn("ciudad", F.initcap("ciudad")) \
.withColumn("region", F.upper("region")) \
.withColumn("estado", F.lower("estado"))

#transformacion de propiedad:
#formatea campos de texto
df_propiedad_clean = df_propiedades.withColumn("tipo", F.lower(F.trim("tipo"))) \
.withColumn("estado", F.lower(F.trim("estado")))
#clasifica las propiedades por su tamanio
df_propiedad_clasificada = df_propiedad_clean.withColumn(
    "categoria_tamanio",
    F.when(F.col("tamanio_m2") < 50, "pequeña")
     .when((F.col("tamanio_m2") >= 50) & (F.col("tamanio_m2") < 100), "mediana")
     .otherwise("grande")
)
#calcula el precio por m2
df_propiedad_precio = df_propiedad_clasificada.withColumn(
    "precio_m2",
    F.when(F.col("tamanio_m2") > 0, F.col("precio") / F.col("tamanio_m2")).otherwise(None)
)
#facilitador de disponibilidad
df_propiedades_procesado = df_propiedad_precio.withColumn(
    "disponible",
    F.when(F.col("estado") == "available", True).otherwise(False)
)
```

```python
#DESNORMALIZACION PROYECTO - PROPIEDAD
# Renombrar columnas del proyecto para que no choquen
df_proyectos_renombrado = df_proyectos_procesado \
    .withColumnRenamed("nombre", "nombre_proyecto") \
    .withColumnRenamed("estado", "estado_proyecto")

# Renombrar 'estado' de propiedad también si lo necesitas
df_propiedades_renombrado = df_propiedades_procesado \
    .withColumnRenamed("estado", "estado_propiedad")

# Join: agregar datos del proyecto a la propiedad
df_propiedad_desnormalizado = df_propiedades_renombrado.join(
    df_proyectos_renombrado,
    on="id_proyecto",  # campo común
    how="left"
).select(
    "id_propiedad",
    "id_proyecto",
    "tipo",
    "tamanio_m2",
    "habitaciones",
    "banios",
    "precio",
    "disponible",
    "precio_m2",
    "categoria_tamanio",
    "estado_propiedad",
    "nombre_proyecto",
    "ciudad",
    "region",
    "fecha_lanzamiento",
    "anio_lanzamiento",
    "estado_proyecto"
)
```

```python
# Registrar tablas como vistas temporales en la sesión actual

try:
    # Registrar tablas como vistas temporales
    df_broker_procesado.createOrReplaceTempView("dim_brokers")
    df_campanias_procesado.createOrReplaceTempView("dim_campanias")
    df_ventas_procesado.createOrReplaceTempView("fact_ventas")
    df_clientes_procesado.createOrReplaceTempView("dim_clientes")
    df_interesados_procesado.createOrReplaceTempView("fact_interesados")
    df_propiedad_desnormalizado.createOrReplaceTempView("dim_propiedades_desnormalizadas")

    # Verificar que se han creado correctamente
    tables = spark.sql("SHOW TABLES").collect()
    print("Tablas disponibles en la sesión:")
    for table in tables:
        print(f" - {table.tableName}")
    
    print("✅ Vistas temporales creadas correctamente para la sesión actual")
    
    # Alternativa: Guardar como tablas directamente en el Lakehouse 
    # Esto registra las tablas en el catálogo del Lakehouse actual
    print("\nRegistrando tablas en el catálogo del Lakehouse...")
    
    # Registrar los dataframes como tablas en el Lakehouse actual parte Tables
    df_broker_procesado.write.format("delta").mode("overwrite").saveAsTable("dim_brokers")
    df_campanias_procesado.write.format("delta").mode("overwrite").saveAsTable("dim_campanias")
    df_ventas_procesado.write.format("delta").mode("overwrite").saveAsTable("fact_ventas")
    df_clientes_procesado.write.format("delta").mode("overwrite").saveAsTable("dim_clientes")
    df_interesados_procesado.write.format("delta").mode("overwrite").saveAsTable("fact_interesados")
    df_propiedad_desnormalizado.write.format("delta").mode("overwrite").saveAsTable("dim_propiedades")
    
    print("✅ Tablas creadas correctamente en el catálogo del Lakehouse")
except Exception as e:
    print(f"Error al crear tablas: {str(e)}")
```
